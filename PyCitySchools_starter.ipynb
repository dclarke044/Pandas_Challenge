```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data = "Resources/schools_complete.csv"
student_data = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_info = pd.read_csv(school_data)
student_info = pd.read_csv(student_data)

# Combine the data into a single dataset
school_info_total = pd.merge(student_info, school_info, how="left", on=["school_name", "school_name"])
school_info_total.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Calculating Totals
total_schools = school_info_total["School ID"].nunique()
total_students = school_info_total["Student ID"].nunique()

# Set data type as integer for budget column, find unique values in budget column, sum
total_budget = school_info_total["budget"].astype(int)
total_budgets = school_info_total["budget"].unique()
total = sum(total_budgets)

# Calculating Averages
avg_math = school_info_total["math_score"].mean()
avg_read = school_info_total["reading_score"].mean()

# Calculating Overall passing rate
pass_rate = (avg_math+avg_read)/2

# Calculating percentage of students passing each subject
# Set passing grade to 70 or higher, count number of students with passing grade, then calc percent
math_pass = school_info_total.loc[school_info_total["math_score"] >= 70]
stud_math = math_pass["Student ID"].count()
per_math = (stud_math/total_students)*100

read_pass = school_info_total.loc[school_info_total["reading_score"] >= 70]
stud_read = read_pass["reading_score"].count()
per_read = (stud_read/total_students)*100

# Create dataframe for District Summary
District_Summary = pd.DataFrame({
    "Total Schools":[total_schools],
    "Total Students": f"{total_students:,}",
    "Total Budget": f"${total:,}",
    "Average Math Score": f"{avg_math:.1f}%",
    "Average Reading Score": f"{avg_read:.1f}%",
    "% Passing Math": f"{per_math:.0f}%",
    "% Passing Reading": f"{per_read:.0f}%",
    "Overall Passing Rate": f"{pass_rate:.0f}%"
})

District_Summary
```

## School Summary

  * Total Students

```python
# Pull and group School Names, etc.
stud_per_school = school_info_total.groupby(['school_name', 'type']).count()['Student ID']
total_per_school = school_info.groupby(['school_name', 'type']).sum()['budget']
total_per_student = total_per_school/stud_per_school
school_avg_math = school_info_total.groupby(['school_name', 'type']).mean()['math_score']
school_avg_read = school_info_total.groupby(['school_name', 'type']).mean()['reading_score']
stud_per_math = school_info_total[school_info_total["math_score"] >= 70].groupby(['school_name', 'type']).count()
school_per_math = (stud_per_math['math_score']/stud_per_school)*100
stud_per_reading = school_info_total[school_info_total["reading_score"] >= 70].groupby(['school_name', 'type']).count()
school_per_reading = (stud_per_reading['reading_score']/stud_per_school)*100
school_pass_rate = (school_per_math + school_per_reading)/2

School_Summary = pd.DataFrame({
    "School Name":{'school_name'},
	"School Type": {'type'},
    "Total Students": f"{stud_per_school}",
    "Total School Budget": f"${total_per_school}",
	"Per Student Budget": f"${total_per_student}",
	"Average Math Score": f"{school_avg_math}%",
	"Average Reading Score": f"{school_avg_read}%",
	"% Passing Math": f"{school_per_math}%",
	"% Passing Reading": f"{school_per_reading}%",
	"Overall Passing Rate": f"{school_pass_rate}%"
})

School_Summary
```

## Math Scores by Grade

```python
# Pull information to calculate grades
math_scores = school_info_total[["school_name", "grade", "math_score"]]
math_scores_9 = math_scores[math_scores["grade"] == "9th"].groupby(["school_name"]).mean()["math_score"]
math_scores_10 = math_scores[math_scores["grade"] == "10th"].groupby(["school_name"]).mean()["math_score"]
math_scores_11 = math_scores[math_scores["grade"] == "11th"].groupby(["school_name"]).mean()["math_score"]
math_scores_12 = math_scores[math_scores["grade"] == "12th"].groupby(["school_name"]).mean()["math_score"]

math_by_grade = pd.DataFrame({"9th Grade":math_scores_9, "10th":math_scores_10, "11th":math_scores_11, "12th":math_scores_12})
math_by_grade
```
